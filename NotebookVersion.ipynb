```python
# Function to reset target values back to 0 and 1
def reset_target(x):
    if x > 0:
        return 1
    return 0
```

```python
# Splitting the dataset into training and testing datasets
X_train, X_test, y_train, y_test = train_test_split(water_df.drop(labels=len(water_df.columns)-1, axis=1),
                                                    water_df.iloc[:, -1], train_size=0.7, random_state=17)
```

```python
# These two modules are developed by me
# and can be found in the repository
from Perceptron import Perceptron
from Metrics import Metrics
```

```python
inp = X_train
inp['target'] = y_train
```

```python
testing = X_test
testing['target'] = y_test
```

```python
my_model = Perceptron(alpha=1e-4, epochs=10, activation='relu')
```

```python
# Training the perceptron model
my_model.train(inp)
```

```python
# testing the perceptron model
result, target = my_model.test(testing)
```

```python
# Metrics has methods to calculate efficiency and confusion matrix
my_metrics = Metrics(target, result)
```

```python
my_metrics.efficiency()
```

```python
my_metrics.confusion_matrix()
```

```python
# n combinations of alpha, epochs and activation have to be provided
n = int(input('Number of combinations: '))
params = []
for i in range(n):
    print(f'Set {i+1}')
    a = float(input('Enter learning rate (alpha):'))
    ep = int(input('Enter number of epochs: '))
    act = input('Enter activation function: ')
    params.append([a, ep, act])
```

```python
max_eff = 0
best_model = None
best_model_metrics = None
```

```python
# n number of peceptron models are created, trained and tested
# Model that gives the maximum efficiency among these, is returned
for i in range(n):
    # curr_model is the perceptron model corresponding to current iteration
    curr_model = Perceptron(alpha=params[i][0], epochs=params[i][1], activation=params[i][2])
    # model is trained
    curr_model.train(inp)
    # model is tested
    res, tar = curr_model.test(testing)
    # metrics
    curr_metrics = Metrics(tar, res)
    curr_eff = curr_metrics.efficiency()
    print(f'Efficiency of Model {i+1}: {curr_eff}')
    # finding the maximum efficiency
    if float(curr_eff) > max_eff:
        max_eff = curr_eff
        best_model = curr_model
        best_model_metrics = curr_metrics
```

```python
print(f'Maximum efficiency is: {max_eff} with alpha={best_model.alpha}, epochs={best_model.epochs}, activation={best_model.getActivationFunction()}')
print(f'___________________________________\n')
print(f'Confusion matrix of the best model')
print(f'___________________________________\n')
print(f'{best_model_metrics.confusion_matrix()}')
print(f'___________________________________\n')
```
