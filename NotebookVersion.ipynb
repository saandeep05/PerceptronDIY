```python
cancer = pd.read_csv("breast-cancer.csv")
```

```python
cancer
```

```python
cancer.drop(labels='id', axis=1, inplace=True)
```

```python
def set_target(x):
    if x == 'M':
        return 1
    return 0
```

```python
cancer['target'] = cancer['diagnosis'].apply(lambda x: set_target(x))
```

```python
cancer.drop(labels='diagnosis', axis=1, inplace=True)
```

```python
cancer
```

```python
scaler2 = StandardScaler()
```

```python
cancer = pd.DataFrame(scaler2.fit_transform(cancer))
```

```python
cancer
```

```python
cancer.iloc[:, -1] = cancer.iloc[:, -1].apply(lambda x : reset_target(x))
```

```python
X_train1, X_test1, y_train1, y_test1 = train_test_split(cancer.drop(labels=len(cancer.columns)-1, axis=1),
                                                    cancer.iloc[:, -1], train_size=0.7, random_state=17)
```

```python
cancer_input = X_train1
cancer_input['target'] = y_train1
```

```python
cancer_testing = X_test1
cancer_testing['target'] = y_test1
```

```python
# an_perceptron = Perceptron(alpha=100, theta=1, epochs=1000)
```

```python
an_perceptron = Perceptron(alpha=1e-5, theta=-0.5, epochs=200)
```

```python
# alpha=1e-5, theta=-0.05, epochs=10 => 78.95
# alpha=1e-5, theta=-0.05, epochs=100 => 95.91
# alpha=1e-5, theta=-0.5, epochs=100 => 97.66 (twice => epochs = 200) (**least false negatives**)
# alpha=1e-5, theta=-0.5, epochs=300 => 95.91
# alpha=1, theta=-0.05, epochs=10 => 95.91
# alpha=1, theta=-0.05, epochs=50 => 96.49 (least false positives)
# alpha=100, theta=1, epochs=50 => 95.91
# alpha=100, theta=1, epochs=100 => 95.91
```

```python
an_perceptron.train(cancer_input)
```

```python
res1, tar1 = an_perceptron.test(cancer_testing)
```
